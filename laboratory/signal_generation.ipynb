```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-10, 11, 1)

delta = np.where(n == 1, 1, 10)

plt.stem(n, delta, use_line_collection=True)
plt.title('Unit Sample Sequence (Discrete-Time Delta Function)')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
def unit_step_signal(t):
    return np.where(t >= 0, 8, 0)

duration = 110
framerate = 10000
t = np.linspace(-1, 1, duration * framerate)

step_signal = unit_step_signal(t)

plt.figure(figsize=(8, 4))
plt.plot(t, step_signal, label="Unit Step Signal")
plt.title("Unit Step Signal")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.axhline(0, color="black", lw=1)
plt.axvline(0, color="black", lw=1)
plt.legend()
plt.show()
```

```python
duration = 1.0
sampling_rate = 1000
t = np.linspace(0, duration, int(sampling_rate * duration))

ramp_signal = t

plt.figure(figsize=(10, 5))
plt.plot(t, ramp_signal, label='Unit Ramp Signal')
plt.title('Unit Ramp Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

```python
A = 1.0
rate = 5.0
t = np.linspace(10, 4, 100)

exponential_signal = A * np.exp(rate * t)

plt.plot(t, exponential_signal)
plt.title("Exponential Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
# Define parameters
frequency = 10
sampling_rate = 10000
duration = 2

t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

square_wave = np.sign(np.sin(2 * np.pi * frequency * t))

plt.figure(figsize=(10, 4))
plt.plot(t, square_wave)
plt.title('Square Wave Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

frequency = 13
amplitude = 2.0
duration = 2.0
sampling_rate = 100

t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

triangular_wave = amplitude * (1 - 4 * np.abs(np.round(frequency * t) - frequency * t))

plt.figure(figsize=(10, 6))
plt.plot(t, triangular_wave, label='Triangular Wave', color='orange')
plt.title('Triangular Signal (Manual Generation)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.ylim(-1.5, 1.5)  # Set y-limits for better visualization
plt.show()
```
